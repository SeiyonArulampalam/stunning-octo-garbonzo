# Fill-in patterns

In the Cholesky factorization, we want to factor the matrix A in the form $A = LL^T$ where $L$ is a lower triangular matrix. Goal is to compute the NZ pattern of L based on the NZ pattern of A. Note, the NZ pattern of L is the NZ pattern of A plus some fill-in entries. The rule for this is:
* $a_{ij} \neq 0 \rightarrow l_{ij} \neq 0$ (copies NZ of A to L)
* $(i < j < k) : \ l_{ji} \neq 0 \land l_{ki} \neq 0 \rightarrow l_{kj} \neq 0$ (path i to j and i to k results in fill-in for kj).

```python
import numpy as np
from _plot_arr import plot_arr

# define array to illustrate fill-in feature
arr = np.diag(-1 * np.ones((5,)))
arr[2,1] = 1.0
arr[4,1] = 1.0
arr[4,2] = 0.5

# rule : on column i if rows j and k NZ => then kj is NZ

# (2,1) & (4,1) => (4,2)

plot_arr(arr)
```

```python
import numpy as np
from _plot_arr import plot_arr
np.random.seed(123456)

# define array to illustrate fill-in feature
N = 8
A = np.diag(0.5 * np.ones((N,)))
for col in range(N):
    for row in range(col+1, N):
        A[row, col] = int(np.random.rand(1)[0] > 0.5)
plot_arr(A)

# fill-in NZ
L = A.copy()
L[7,3] = -1.0
L[7,6] = -1.0
plot_arr(L)
```
